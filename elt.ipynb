# ELT Project: GDP and World Happiness Report 


```python
# Import Depenencies
import pandas as pd
import requests
```

# Extraction

```python
# Read in csv of World Happiness Report 2018
cvs_file="2018.csv"
happiness_df = pd.read_csv (cvs_file)
happiness_df.head()
```

```python
# Read in csv of GDP
csv_file="GDP.csv"
gdp_df= pd.read_csv(csv_file)
gdp_df.head()
```

# Transformation

```python
# Display columns in happiness report
happiness_df.columns
```

```python
# Drop the unwanted columns
happiness_df=happiness_df.drop(columns =['Social support',
                                         'Healthy life expectancy', 
                                         'Freedom to make life choices', 
                                         'Generosity', 
                                         'Perceptions of corruption'], axis=1)
happiness_df.head()
```

```python
# Rename country column
happiness_df=happiness_df.rename(columns={'Country or region':'Country'})
```

```python
# Remove duplicates for World Happiness Report
happiness_df=happiness_df.drop_duplicates(keep='first')
happiness_df.head()
```

```python
# show all columns for GDP
gdp_df.columns
```

```python
# Remove all unneccessary columns
gdp_df=gdp_df.drop(columns =['1990','1991','1992','1993','1994',
                            '1995','1996','1997','1998','1999','2000', 
                            '2001','2002','2003','2004','2005','2006', 
                            '2007', '2008','2009','2010','2011','2012',
                            '2013','2014','2015','2016','2017','2019'], axis=1)
gdp_df.head(100)
```

```python
# Remove duplicates for GDP
gdp_df=gdp_df.drop_duplicates(keep='first')
gdp_df.head()
```

```python
# see gdp columns
gdp_df.columns
```

```python
# Rename column in GDP
gdp_df=gdp_df.rename(columns={'Country ':'Country',
                              '2018': 'GDP 2018'})
```

```python
# Merging the dataframes together 
merge_df=pd.merge(gdp_df,happiness_df,on='Country',how='inner')
merge_df.head()
```

```python
# Rearranging the columns and overall rank to ascending order
clean_df=merge_df[['Overall rank','Country','Country Code','Score','GDP 2018','GDP per capita']].sort_values(by="Overall rank",ascending=True)
clean_df.head()
```

```python
# Changing column names for loading section
clean_df=clean_df.rename(columns={'Country':'country',
                              'Overall rank': 'overall_rank',
                                 'Country Code': 'country_code',
                                 'GDP 2018':'gdp_2018',
                                 'Score':'score',
                                 'GDP per capita':'gdp_per_capita'})
clean_df.head()
```
